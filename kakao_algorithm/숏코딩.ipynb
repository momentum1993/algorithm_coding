# 숏코딩

[2018 카카오 코드 페스티벌 '숏코딩' 문제 링크](https://www.acmicpc.net/problem/15956)  

### 문제    
코드 페스티벌 온라인 예선에 참가하고 있던 라이언은 이제 남은 시간이 02:07:19밖에 없다는 것을 깨닫게 되었다. 라이언은 이미 머릿속에서 풀이를 구상하고 코딩도 완료했기 때문에, 이를 그대로 타이핑하기만 하면 된다.  
  
![](https://upload.acmicpc.net/6d222569-3f51-447f-9ea9-9a9b77bfe041/-/preview/)  
  
지금 라이언은 변수들과 정수들끼리 같은지 다른지 비교하는 간단한 조건문 (conditional expression) S를 작성하고자 한다. 자세히 설명하자면,  
  
- 라이언이 작성하는 변수의 이름은 영문 알파벳으로만 구성된 문자열이다. 예를 들어, kakao, COde, festival은 변수의 이름이 될 수 있지만, C0de, 2018, hello_world는 변수의 이름이 될 수 없다. 변수는 정수 값을 저장하고 있으며, 변수의 값은 해당 변수가 저장하고 있는 정수 값을 의미한다.  
- 라이언이 작성하는 정수는 무조건 -109 이상 109 이하이다. 0을 제외한 정수는 숫자 0으로 시작하지 않으며, 0을 표기하는 유일한 방법은 0이다. 예를 들어, 라이언은 0, 123456, -987654321 과 같은 정수는 작성할 수 있지만, -0, 0123, 00000, -009876, 999999999999999와 같은 정수는 작성할 수 없다.  
- 단항식은 변수 또는 정수이다.  
- 라이언은 아래 두 개의 비교 연산자를 사용하여 논리식을 만든다.  
    - 같음 연산자 ==: <a\>==<b\>와 같이 사용하며, 두 단항식 <a\>와 <b\>의 값을 계산하여, 같으면 true를, 다르면 false를 반환한다.  
    - 다름 연산자 !=: <a\>!=<b\>와 같이 사용하며, 두 단항식 <a\>와 <b\>의 값을 계산하여, 다르면 true를, 같으면 false를 반환한다.  
    - 위의 두 연산자 모두, 연산자의 좌변과 우변에는 오직 단항식만이 올 수 있다. 예를 들어, (a==b)!=(b==c)는 !=의 좌변과 우변에 단항식이 아닌 논리식이 들어 있으므로 라이언이 사용하는 논리식이 아니다.  
- 라이언은 논리곱 연산자 &&로 한 개 이상의 논리식들을 연결하여 조건문을 만든다. 연결한 논리식들이 모두 true를 반환할 때에만 true, 그렇지 않다면 false를 반환하도록 할 것이다. 연결할 때에 공백 등을 삽입하지 않는다. 예를 들어, 라이언이 세 논리식 a==b, b!=c, 1!=a를 연결하고자 한다면, a==b&&b!=c&&1!=a와 같이 연결한다.  
  
라이언이 작성하고자 하는 조건문 S는 위 문단에서 언급한 방법으로만 만들 수 있음이 보장된다.  
  
하지만, 라이언은 남은 시간에 비해 S의 길이가 너무 길다고 생각하여, S와 동치이면서 위에서 제시된 방법으로 만들 수 있는 길이가 가장 짧은 조건문 S'을 대신 입력할 생각이다. 두 조건문 T1과 T2가 동치라는 것은, 어떤 변수 (조건문에 등장하지 않는 모든 가능한 변수도 포함) 에 어떤 값 (라이언이 작성하지 않는 정수도 가능)을 넣더라도 T1의 참/거짓 여부와 T2의 참/거짓 여부가 같다는 것이다.  
  
라이언을 위해, S'을 대신 구해주는 프로그램을 작성하자. S'이 여러 가지 있다면, 그 중 하나를 아무거나 구해주면 된다.  
  
### 입력  
첫 번째 줄에 라이언이 입력하고자 하는 조건문 S(1 ≤ |S| ≤ 106)가 주어진다.  
  
### 출력  
첫 번째 줄에 S와 동치이면서, 문제 본문에 제시된 방법으로 만들 수 있는 길이가 가장 짧은 조건문 S'을 출력한다.    

```python
while True:
    string = input("input: ") # 입력
    if string == "exit": 
        print("exit the program")
        break # 프로그램 종료
    word = {} # 각 단어별 인덱스
    val_per_idx = {} # 인덱스별 가지는 변수의 값
    word_per_idx = {} # 인덱스별로 똑같은 값을 지니는 변수 리스트
    diff = [] # != 연산을 적용할 피연산자들

    expressions = string.split("&&") # && 기준으로 나눔
    answer = '' # 답
    idx = 1 # 인덱스
    for e in expressions: # &&기준으로 나눈 연산식들에 대해서
        if '==' in e: # == 연산
            key,val = tuple(e.split("==")) # 피연산자1, 피연산자2
            key_idx = word.get(key) # 피연산자1의 인덱스
            val_idx = word.get(val) # 피연산자2의 인덱스
            if key_idx and val_idx: # 만약 둘다 이미 나온 피연산자라면
                for k,v in word.items():
                    if k == key or k == val:
                        word[k] = key_idx # 피연산자1의 인덱스로 모두 통일
            elif not key_idx and val_idx: # 피연산자2만 이미 나왔다면
                word[key] = val_idx # 피연산자1의 인덱스를 피연산자2로 할당
            elif key_idx and not val_idx: # 피연산자1만 이미 나왔다면
                word[val] = key_idx # 피연산자2의 인덱스를 피연산자1로 할당
            else: # 두 피연산자가 처음 나오는 변수들이라면
                word[key] = idx
                word[val] = idx
                # 동일한 인덱스 새로 할당

            if val.isdigit() or val[1:].isdigit():
            # 피연산자2가 변수가 아닐 경우
                val_per_idx[word[val]] = val
                # 피연산자1의 인덱스에 대해 변수값을 저장
                del(word[val])

        elif '!=' in e: # != 연산
            key,val = tuple(e.split("!="))
            if not word.get(key):
                word[key] = idx
            if not word.get(val):
                word[val] = idx + 1
            # 연산자에 인덱스 할당
            diff.append((word[key], word[val]))
            # 그리고 != 연산할 (피연산자1, 피연산자2) 저장
        idx += 2 # 인덱스 증가

    for k,v in word.items():
        if word_per_idx.get(v):
            word_per_idx[v].append(k)
        else:
            word_per_idx[v] = [k]
    # 동일한 인덱스를 지니는 단어들을 함께 저장

    for k, v in word_per_idx.items():
        words = sorted(v, key=lambda x:len(x))
        # 동일한 인덱스를 지닌 단어들을 길이 오름차순으로 정렬
        idx = k # 인덱스
        val = words[-1]
        if val_per_idx.get(k): # 만약 변수값이 있다면
            val = val_per_idx[k]
            answer += "&&" + words[0] + "==" + val
            # 길이가 가장 작은 변수에 변수값 할당
        if len(words[0]) > len(val):
        # 할당되는 값의 길이와 최소 길이 이름의 변수의 길이 비교
            for w in words[1:]:
                answer += "&&" + w + "==" + val
        else:
            for w in words[1:]:
                answer += "&&" + words[0] + "==" + w
        # 더 길이가 짧은 값에 대해서 답을 붙임

    for a, b in diff: # != 연산에 대해서
        left = sorted(word_per_idx[a], key=lambda x:len(x))[0]
        # a 인덱스를 지니는 변수 중에 이름 길이가 최소인 변수
        right = sorted(word_per_idx[b], key=lambda x:len(x))[0]
        # b 인덱스를 지니는 변수 중에 이름 길이가 최소인 변수
        if val_per_idx.get(b):
            if len(right) > len(val_per_idx[b]):
                right = val_per_idx[b]
        # right에 right와 변수값 중 길이가 작은 값을 할당
        answer += "&&" + left + "!=" + right

    print("answer: " + answer[2:]) # 답 출력
    print()
```

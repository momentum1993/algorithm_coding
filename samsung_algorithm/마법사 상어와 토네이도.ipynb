# 마법사 상어와 토네이도

[baekjoon의 '마법사 상어와 토네이도' 링크](https://www.acmicpc.net/problem/20057)

### 문제  
마법사 상어가 토네이도를 배웠고, 오늘은 토네이도를 크기가 N×N인 격자로 나누어진 모래밭에서 연습하려고 한다. 위치 (r, c)는 격자의 r행 c열을 의미하고, A[r][c]는 (r, c)에 있는 모래의 양을 의미한다.  
  
토네이도를 시전하면 격자의 가운데 칸부터 토네이도의 이동이 시작된다. 토네이도는 한 번에 한 칸 이동한다. 다음은 N = 7인 경우 토네이도의 이동이다.  
  
![](https://upload.acmicpc.net/37e7aa13-0f2b-49d6-af68-e745537b1ea3/-/preview/)  
  
토네이도가 한 칸 이동할 때마다 모래는 다음과 같이 일정한 비율로 흩날리게 된다.  
  
![](https://upload.acmicpc.net/33b01ca0-4659-49f1-b126-8e042e17d3f1/-/preview/)  
  
토네이도가 x에서 y로 이동하면, y의 모든 모래가 비율과 α가 적혀있는 칸으로 이동한다. 비율이 적혀있는 칸으로 이동하는 모래의 양은 y에 있는 모래의 해당 비율만큼이고, 계산에서 소수점 아래는 버린다. α로 이동하는 모래의 양은 비율이 적혀있는 칸으로 이동하지 않은 남은 모래의 양과 같다. 모래가 이미 있는 칸으로 모래가 이동하면, 모래의 양은 더해진다. 위의 그림은 토네이도가 왼쪽으로 이동할 때이고, 다른 방향으로 이동하는 경우는 위의 그림을 해당 방향으로 회전하면 된다.  
  
토네이도는 (1, 1)까지 이동한 뒤 소멸한다. 모래가 격자의 밖으로 이동할 수도 있다. 토네이도가 소멸되었을 때, 격자의 밖으로 나간 모래의 양을 구해보자.  
  
### 입력  
첫째 줄에 격자의 크기 N이 주어진다. 둘째 줄부터 N개의 줄에는 격자의 각 칸에 있는 모래가 주어진다. r번째 줄에서 c번째 주어지는 정수는 A[r][c] 이다.  
  
### 출력  
격자의 밖으로 나간 모래의 양을 출력한다.  
  
### 제한  
- 3 ≤ N ≤ 499
- N은 홀수
- 0 ≤ A[r][c] ≤ 1,000
- 가운데 칸에 있는 모래의 양은 0  

```python
from collections import defaultdict

def tornado_simulation(N, sand, y, x): # 토네이도 진행 시뮬레이션
    answer = 0 # 답
    direction = [(0, -1), (1, 0), (0, 1), (-1, 0)] # 서, 남, 동, 북
    scattered_sand = [[(1, -1, 1), (1, 1, 1), (2, -2, 0), (2, 2, 0), (5, 0, -2), (7, -1, 0), (7, 1, 0), (10, -1, -1), (10, 1, -1)], # 토네이도가 서쪽으로 향했을 때(= 동쪽에서 왔을 때)
                  [(1, -1, -1), (1, -1, 1), (2, 0, -2), (2, 0, 2), (5, 2, 0), (7, 0, -1), (7, 0, 1), (10, 1, -1), (10, 1, 1)], # 남(= 북)
                  [(1, -1, -1), (1, 1, -1), (2, -2, 0), (2, 2, 0), (5, 0, 2), (7, -1, 0), (7, 1, 0), (10, -1, 1), (10, 1, 1)], # 동(= 서)
                  [(1, 1, -1), (1, 1, 1), (2, 0, -2), (2, 0, 2), (5, -2, 0), (7, 0, -1), (7, 0, 1), (10, -1, -1), (10, -1, 1)]] # 북(= 남)
    move = 1 # 현재의 방향에서 이동해야하는 칸 
    count = 0 # 움직이는 칸의 수가 N-1 만큼 될 때까지 count가 2개가 될 때마다 움직이는 칸 수를 1 증가시킨다.
    # ex) move -> 1 1 2 2 3 3 4 4 ... 
    d = 0 # 현재 토네이도가 이동하는 방향
    while y != 0 or x != 0: # (y, x) == (0, 0) 이 될 때까지
        add_y, add_x = direction[d] # d방향으로의 움직일 좌표
        for _ in range(move): # move칸 만큼 d방향으로 토네이도가 이동한다
            if y+add_y < 0 or y+add_y >= N or x+add_x < 0 or x+add_x >= N: # 보드를 벗어난다면 탐색 중지
                break
            y += add_y
            x += add_x
            # d방향으로 1칸 이동
            
            s = sand[y][x] # 이동 후에 도달한 칸에 있는 모래의 양
            sand[y][x] = 0 # 해당 칸의 모래는 토네이도로 인해 모두 쓸려나간다.
            scattered = 0 # 흩뿌려진 모래 양
            for p, m_y, m_x in scattered_sand[d]: # 해당 방향에 따른 흩어질 모래에 대해서 
            # (p: 퍼센티지, m_y: 현재좌표와의 y좌표 차이, m_x: 현재좌표와의 x좌표 차이)
                if 0 <= y+m_y < N and 0 <= x+m_x < N: # 보드 내의 범위라면
                    sand[y+m_y][x+m_x] += int(s * (p / 100)) # 해당 퍼센티지의 모래를 좌표로 보낸다. 
                else: # 보드 밖의 범위라면
                    answer += int(s * (p / 100)) # 답에 보드 밖에 흩뿌려진 모래의 양 추가
                scattered += int(s * (p / 100)) # 흩뿌려진 모래의 양 추가
                
            # 흩뿌려지고 남은 모래에 대해서
            if 0 <= y+add_y < N and 0 <= x+add_x < N: # 현재 방향으로 1칸 앞이 보드 내의 범위라면
                sand[y+add_y][x+add_x] += s - scattered # 남은 모래를 이동시킨다.
            else: # 보드 밖이라면
                answer += s - scattered # 답에 남은 모래의 양을 추가한다.
        count += 1 # count 증가
        if move < N-1 and count == 2: # 만약 count가 2이고, move가 N-1보다 작다면
            move += 1 # 토네이도가 특정 방향으로 한번에 움직이는 칸 move 1 증가
            count = 0
        d += 1 
        d %= 4
        # move칸 만큼 토네이도가 휩쓸고 간 후, 다시 방향 전환해준다.
    return answer # 답 출력
    
T = int(input("Testcase num: "))
for _ in range(T):
    print("\n\nInput")
    N = int(input()) # 보드의 너비 및 높이
    sand = [] # 보드위의 각 좌표별 모래의 양
    for _ in range(N):
        sand.append(list(map(int, input().split())))
    print("\nOutput")
    print(tornado_simulation(N, sand, N//2, N//2)) # 시뮬레이션 진행 후 답 출력
```

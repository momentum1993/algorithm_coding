### 문제  
섬으로 이루어진 나라가 있고, 모든 섬을 다리로 연결하려고 한다. 이 나라의 지도는 N×M 크기의 이차원 격자로 나타낼 수 있고, 격자의 각 칸은 땅이거나 바다이다.  
  
섬은 연결된 땅이 상하좌우로 붙어있는 덩어리를 말하고, 아래 그림은 네 개의 섬으로 이루어진 나라이다. 색칠되어있는 칸은 땅이다.  
  
![](https://upload.acmicpc.net/38cb578e-b289-4b72-841e-422a1458d617/-/preview/)  
  
다리는 바다에만 건설할 수 있고, 다리의 길이는 다리가 격자에서 차지하는 칸의 수이다. 다리를 연결해서 모든 섬을 연결하려고 한다. 섬 A에서 다리를 통해 섬 B로 갈 수 있을 때, 섬 A와 B를 연결되었다고 한다. 다리의 양 끝은 섬과 인접한 바다 위에 있어야 하고, 한 다리의 방향이 중간에 바뀌면 안된다. 또, 다리의 길이는 2 이상이어야 한다.  
  
다리의 방향이 중간에 바뀌면 안되기 때문에, 다리의 방향은 가로 또는 세로가 될 수 밖에 없다. 방향이 가로인 다리는 다리의 양 끝이 가로 방향으로 섬과 인접해야 하고, 방향이 세로인 다리는 다리의 양 끝이 세로 방향으로 섬과 인접해야 한다.  
  
섬 A와 B를 연결하는 다리가 중간에 섬 C와 인접한 바다를 지나가는 경우에 섬 C는 A, B와 연결되어있는 것이 아니다.  
  
아래 그림은 섬을 모두 연결하는 올바른 2가지 방법이고, 다리는 회색으로 색칠되어 있다. 섬은 정수, 다리는 알파벳 대문자로 구분했다.

나라의 정보가 주어졌을 때, 모든 섬을 연결하는 다리 길이의 최솟값을 구해보자.  
  
### 입력
첫째 줄에 지도의 세로 크기 N과 가로 크기 M이 주어진다. 둘째 줄부터 N개의 줄에 지도의 정보가 주어진다. 각 줄은 M개의 수로 이루어져 있으며, 수는 0 또는 1이다. 0은 바다, 1은 땅을 의미한다.    
  
### 출력  
모든 섬을 연결하는 다리 길이의 최솟값을 출력한다. 모든 섬을 연결하는 것이 불가능하면 -1을 출력한다.  
  
### 제한  
- 1 ≤ N, M ≤ 10  
- 3 ≤ N×M ≤ 100  
- 2 ≤ 섬의 개수 ≤ 6  
  

```python
from queue import Queue
from queue import PriorityQueue

def find_island(N, M, board, init_y, init_x, island_idx): # 섬을 구분짓는 함수
    direction = [(1, 0), (-1, 0), (0, 1), (0, -1)] # 네 방향
    q = Queue() # 큐
    q.put((init_y, init_x)) # 큐에 (init_y, init_x)정보를 push하고 해당 좌표에 해당하는 섬의 영토를 찾을 것이다.
    coastline = set() # 현재 탐색하는 섬의 해안선에 해당하는 좌표들
    while not q.empty():
        y, x = q.get()
        board[y][x] = island_idx # 섬에 해당하는 좌표에는 섬의 인덱스를 저장
        coast = False # 현재 탐색하는 위치가 해안인지의 여부
        for add_y,add_x in direction: # 인접한 네 방향에 대해서
            if 0 <= y+add_y < N  and 0 <= x+add_x < M: # 지도 안의 지역의 경우
                if board[y+add_y][x+add_x] == 1: # 육지라면
                    q.put((y+add_y, x+add_x)) # 해당 좌표를 섬의 영토에 추가해서 탐색
                elif board[y+add_y][x+add_x] == 0: # 바다일 경우
                    coast = True # 현재 좌표는 해안이다.
        if coast:
            coastline.add((y, x)) # 해안선에 해당하는 좌표 추가
    return list(coastline) # 해안선을 이루는 좌표들 반환
    
def build_bridge(N, M, board, coast_line, bridge): # 특정 섬간에 다리를 건설할 수 있는 최소 비용을 찾는 함수
    direction = [(1, 0), (-1, 0), (0, 1), (0, -1)]
    for i in range(len(coast_line)): # 모든 섬 중에
        for y, x in coast_line[i]: # 현재 탐색하는 섬의 해안선에 포함되는 좌표의
            for add_y, add_x in direction: # 인접한 네 방향에 대해서
                if 0 <= y+add_y < N  and 0 <= x+add_x < M and board[y+add_y][x+add_x] == 0: # 지도 안의 영토이고, 바다인 경우
                    # 해당 방향으로 다리를 이어본다.
                    now_y, now_x = add_y+y, add_x+x
                    dist = 0 # 다리 건설 거리 및 비용
                    while 0 <= now_y < N  and 0 <= now_x < M: # 지도 안의 영토일 경우
                        if board[now_y][now_x] > 1 and board[now_y][now_x] != i+2: # 만약 다리를 확장하다가 다른 섬을 만난다면
                            if dist >= 2: # 그리고 그 다리의 길이가 2 이상일 경우
                                bridge[i][board[now_y][now_x]-2] = min(bridge[i][board[now_y][now_x]-2], dist)
                                bridge[board[now_y][now_x]-2][i] = min(bridge[board[now_y][now_x]-2][i], dist)
                                # 다리 건설 비용의 최소인지 확인하고, 최솟값일 경우 저장
                            break # 탐색 중지 후 다른 해안선 좌표에 대해서 탐색
                        dist += 1 # 다리 건설 비용 1 추가
                        now_y += add_y
                        now_x += add_x
                        # 현재의 방향으로 다리 1칸 건설
    return

def cal_minval(bridge): # 섬간에 다리를 건설해서 모든 섬을 이을 수 있도록 할 수 있는 건설 비용 최솟값 반환
    answer = 0 # 답
    connected = [False for _ in range(len(bridge))] # 현재 섬이 연결되었는지 여부
    connected[0] = True # 0번(board상에는 2번) 섬부터 연결을 시작할 것이기 때문에 연결되었음을 체크
    q = PriorityQueue() # 우선순위 큐
    for i in range(1, len(bridge)): # 0번 섬과 연결된 섬에 대해서
        if bridge[0][i] == 11:
            continue
        q.put((bridge[0][i], i)) # 우선순위 큐에 (건설 비용, 연결될 섬) 정보를 push
    while not q.empty(): # 큐가 빌때까지
        dist, island = q.get() # (건설 비용, 연결될 섬) 정보를 pop
        if connected[island]: # 만약 이미 연결된 섬이라면 다음 정보에 대해서 탐색
            continue
        connected[island] = True # 해당 섬이 연결되었음을 표시
        answer += dist # 그리고 건설 비용을 더해준다.
        for i in range(len(bridge)): # 방금 연결된 섬에 대해서
            if connected[i] or bridge[island][i] == 11:
                continue
            q.put((bridge[island][i], i)) # 연결할 수 있는 섬에 대한 (건설 비용, 연결할 수 있는 섬) 정보를 push
    if False in connected: # 모든 섬 연결 시뮬레이션 진행 후에 연결되지 않은 섬이 있다면
        return -1 # -1을 반환
    return answer # 섬이 모두 연결되었다면 최솟값 반환
        

def simulation(N, M, board): # 섬 연결 및 최소비용 탐색 시뮬레이션
    coast_line = [] # 해안선
    island_idx = 2 # 섬의 인덱스
    for y in range(N):
        for x in range(M): # 모든 보드 위의 좌표에 대해서
            if board[y][x] == 1: # 아직 탐색하지 않은 미지의 섬이라면
                coast_line.append(find_island(N, M, board, y, x, island_idx)) # 해당 섬에 인덱스를 표시하고, 해당섬의 해안선 좌표를 얻는다
                island_idx += 1
    bridge = [[11 for _ in range(island_idx-2)] for _ in range(island_idx-2)] # 각 섬끼리 다리 건설 비용의 최솟값
    build_bridge(N, M, board, coast_line, bridge) # 특정 섬끼리 다리 건설 비용의 최솟값 탐색
    return cal_minval(bridge) # 그리고 모든 섬을 연결했을 때의 최솟값을 반환
    

T = int(input("Testcase num: ")) # 테스트케이스 개수
for _ in range(T):
    print("\n\nInput")
    N, M = map(int, input().split()) # 보드의 세로, 가로 길이
    board = [] # 보드
    for _ in range(N):
        board.append(list(map(int, input().split())))
        
    print("\nOutput")
    print(simulation(N, M, board)) # 시뮬레이션 진행 후 최소 다리 길이 출력
```

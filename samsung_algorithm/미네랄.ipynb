# 미네랄

[baekjoon의 '미네랄' 링크](https://www.acmicpc.net/problem/2933)

### 문제   
창영과 상근은 한 동굴을 놓고 소유권을 주장하고 있다. 두 사람은 막대기를 서로에게 던지는 방법을 이용해 누구의 소유인지를 결정하기로 했다. 싸움은 동굴에서 벌어진다. 동굴에는 미네랄이 저장되어 있으며, 던진 막대기가 미네랄을 파괴할 수도 있다.  
  
동굴은 R행 C열로 나타낼 수 있으며, R×C칸으로 이루어져 있다. 각 칸은 비어있거나 미네랄을 포함하고 있으며, 네 방향 중 하나로 인접한 미네랄이 포함된 두 칸은 같은 클러스터이다.  
  
창영은 동굴의 왼쪽에 서있고, 상근은 오른쪽에 서있다. 두 사람은 턴을 번갈아가며 막대기를 던진다. 막대를 던지기 전에 던질 높이를 정해야 한다. 막대는 땅과 수평을 이루며 날아간다.  
  
막대가 날아가다가 미네랄을 만나면, 그 칸에 있는 미네랄은 모두 파괴되고 막대는 그 자리에서 이동을 멈춘다.  
  
미네랄이 파괴된 이후에 남은 클러스터가 분리될 수도 있다. 새롭게 생성된 클러스터가 떠 있는 경우에는 중력에 의해서 바닥으로 떨어지게 된다. 떨어지는 동안 클러스터의 모양은 변하지 않는다. 클러스터는 다른 클러스터나 땅을 만나기 전까지 게속해서 떨어진다. 클러스터는 다른 클러스터 위에 떨어질 수 있고, 그 이후에는 합쳐지게 된다.  
  
동굴에 있는 미네랄의 모양과 두 사람이 던진 막대의 높이가 주어진다. 모든 막대를 던지고 난 이후에 미네랄 모양을 구하는 프로그램을 작성하시오.  
  
### 입력  
첫째 줄에 동굴의 크기 R과 C가 주어진다. (1 ≤ R,C ≤ 100)  
  
다음 R개 줄에는 C개의 문자가 주어지며, '.'는 빈 칸, 'x'는 미네랄을 나타낸다.  
  
다음 줄에는 막대를 던진 횟수 N이 주어진다. (1 ≤ N ≤ 100)  
  
마지막 줄에는 막대를 던진 높이가 주어지며, 공백으로 구분되어져 있다. 모든 높이는 1과 R사이이며, 높이 1은 행렬의 가장 바닥, R은 가장 위를 의미한다. 첫 번째 막대는 왼쪽에서 오른쪽으로 던졌으며, 두 번째는 오른쪽에서 왼쪽으로, 이와 같은 식으로 번갈아가며 던진다.  
  
공중에 떠 있는 미네랄 클러스터는 없으며, 두 개 또는 그 이상의 클러스터가 동시에 떨어지는 경우도 없다. 클러스터가 떨어질 때, 그 클러스터 각 열의 맨 아래 부분 중 하나가 바닥 또는 미네랄 위로 떨어지는 입력만 주어진다.  
  
### 출력  
입력 형식과 같은 형식으로 미네랄 모양을 출력한다. 

```python
from copy import deepcopy

def move_cluster(R, C, board, cp_board, checked, cp_checked, y, x, distance): # 미네랄 클러스터를 distance만큼 바닥쪽으로 옮기는 함수
    direction = [(0,1), (1,0), (0,-1), (-1,0)] # 4방향
    checked[y+distance][x] = True
    board[y+distance][x] = 'x'
    checked[y][x] = False
    board[y][x] = '.'
    # (y,x)의 미네랄을 (y+distance, x) 위치로 옮기고, 탐색했음을 체크한다.
    for add_y, add_x in direction: # 4 방향에 대해서
        if 0 <= add_y+y < R and 0 <= add_x+x < C and cp_board[add_y+y][add_x+x] == 'x' and not cp_checked[add_y+y][add_x+x]:
        # 같은 클러스터 내의 아직 탐색하지 않은 미네랄에 대해서
            cp_checked[add_y+y][add_x+x] = True # 탐색했음을 체크
            move_cluster(R, C, board, cp_board, checked, cp_checked, add_y+y, add_x+x, distance)
            # 재귀적으로 같은 클러스터의 미네랄을 찾고, 이동시킨다.
    return

def search_cluster(R, C, board, checked, y, x, cluster_location): # 같은 클러스트 내에있는 미네랄을 찾고, 옮겨야 하는 높이를 반환
    distance = R #  미네랄을 이동시켜야 하는 높이
    if cluster_location[x] < y: # x좌표별로 현재 탐색하는 미네랄의 위치가 가장 바닥과 가까울 경우
        cluster_location[x] = y # 값을 갱신해준다
        if y == R-1: # 그리고 바닥에 붙어 있을 경우
            distance = 0 # 이동거리는 0
        elif board[y+1][x] != 'x': # 공중에 떠 있는 경우
            temp = 1
            for h in range(y+2, R):
                if board[h][x] == 'x':
                    break
                temp += 1
            distance = temp
        # 다른 클러스터와의 거리나, 바닥과의 거리 중에 가장 짧은 거리를 이동거리로 갱신
    direction = [(0,1), (1,0), (0,-1), (-1,0)]
    for add_y, add_x in direction: # 4 방향에 대해서
        if 0 <= add_y+y < R and 0 <= add_x+x < C and board[add_y+y][add_x+x] == 'x' and not checked[add_y+y][add_x+x]:
        # 같은 클러스터 내의 아직 탐색하지 않은 미네랄 조각에 대해서
            checked[add_y+y][add_x+x] = True # 탐색했음을 체크
            distance = min(distance, search_cluster(R, C, board, checked, add_y+y, add_x+x, cluster_location))
            # 그리고 재귀적으로 같은 클러스터 내의 미네랄 조각을 찾는다.
            # 그리고 최소 이동거리를 반환
    return distance # 현재 미네랄 클러스터의 이동거리 반환
            

R, C = tuple(int(x) for x in input().split()) # 행, 열
board = [] # 보드
for _ in range(R):
    board.append(list(input())) # 미네랄의 상태 입력받음
N = int(input()) # 막대 개수
throw_height = [R - int(x) for x in input().split()] # 막대의 높이

right = False # 막대를 던지는 방향(True: 오른쪽, False: 왼쪽)

for h in throw_height: # 막대를 던지는 모든 시행에 대해서
    checked = [[False for _ in range(C)] for _ in range(R)] # 해당하는 미네랄 좌표를 탐색했는지 여부
    if right: # 오른쪽에서 던짐
        right = False
        for x in range(C-1, -1, -1):
            if board[h][x] == 'x':
                board[h][x] = '.'
                break
        # 높이 h에서 막대를 던졌을 때 맞는 가장 오른쪽에 있는 미네랄을 없애준다.
    else: # 왼쪽에서 던짐
        right = True
        for x in range(C):
            if board[h][x] == 'x':
                board[h][x] = '.'
                break
        # 높이 h에서 막대를 던졌을 때 맞는 가장 왼쪽에 있는 미네랄을 없애준다.
    for y in range(R-1, -1, -1):
        for x in range(C-1, -1, -1): # 보드의 모든 좌표에 대해서
            if board[y][x] == 'x' and not checked[y][x]: # 아직 탐색하지 않은 미네랄 조각이 있다면 
                cluster_location = [-1 for _ in range(C)] # 해당하는 미네랄 클러스터에서 x좌표별 가장 바닥과 가까운 미네랄의 위치
                checked[y][x] = True # (y, x) 좌표 위치의 미네랄을 탐색했음을 체크
                cp_checked = deepcopy(checked) # 현재 상태의 탐색 여부 깊은 복사
                cp_board = deepcopy(board) # 현재 보드 상태 깊은 복사
                distance = search_cluster(R, C, board, checked, y, x, cluster_location)
                # 현재 발견한 미네랄 조각의 클러스터를 탐색하고, 클러스터가 바닥으로 옮겨질 높이 반환
                if distance > 0: # 만약 클러스터가 옮겨져야 한다면
                    move_cluster(R, C, board, cp_board, checked, cp_checked, y, x, distance)
                    # distance 높이 만큼 해당 클러스터를 옮긴다.
    
for y in range(R):
    for x in range(C):
        print(board[y][x], end="")
    print()
# 모든 막대를 던져서 미네랄을 없앤 후 마지막 미네랄 상태를 출력
```

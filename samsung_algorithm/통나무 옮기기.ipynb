### 문제   
가로와 세로의 길이가 같은 평지에서 벌목을 한다. 그 지형은 0과 1로 나타나 있다. 1은 아직 잘려지지 않은 나무를 나타내고 0은 아무 것도 없음을 나타낸다. 다음 지형을 보자.  

~~~ 
B 0 0 1 1
B 0 0 0 0
B 0 0 0 0
1 1 0 0 0
E E E 0 0
~~~

위의 지형에서 길이 3인 통나무 BBB를 밀거나 회전시켜 EEE의 위치로 옮기는 작업을 하는 문제를 생각해 보자. BBB와 EEE의 위치는 임의로 주어진다. 단 문제에서 통나무의 길이는 항상 3이며 B의 개수와 E의 개수는 같다. 통나무를 움직이는 방법은 아래와 같이 상하좌우(Up, Down, Left, Right)와 회전(Turn)이 있다.  
  
|코드	|의미|
|----|----|
|U|	통나무를 위로 한 칸 옮긴다.|
|D|	통나무를 아래로 한 칸 옮긴다.|
|L|	통나무를 왼쪽으로 한 칸 옮긴다.|  
|R|	통나무를 오른쪽으로 한 칸 옮긴다.|  
|T|	중심점을 중심으로 90도 회전시킨다.|

이와 같은 방식으로 이동시킬 때에 그 움직일 위치에 다른 나무, 즉 1이 없어야만 움직일 수 있다. 그리고 움직임은 위의 그림과 같이 한 번에 한 칸씩만 움직인다. 단 움직이는 통나무는 어떤 경우이든지 중간단계에서 한 행이나 한 열에만 놓일 수 있다. 예를 들면 아래 그림에서 a와 같은 단계는 불가능하다. 그리고 회전의 경우에서는 반드시 중심점을 중심으로 90도 회전을 해야 한다. (항상 통나무의 길이가 3이므로 중심점이 있음)  
  
그리고 이런 회전(Turn)이 가능하기 위해서는 그 통나무를 둘러싸고 있는 3\*3 정사각형의 구역에 단 한 그루의 나무도 없어야만 한다. 즉, 아래그림 b, d와 같이 ?로 표시된 지역에 다른 나무, 즉 1이 없어야만 회전시킬 수 있다. 따라서 c와 같은 경우에, 통나무는 왼쪽 아직 벌채되지 않은 나무 때문에 회전시킬 수 없다.

문제는 통나무를 5개의 기본동작(U, D, L, R, T)만을 사용하여 처음위치(BBB)에서 최종위치(EEE)로 옮기는 프로그램을 작성하는 것이다. 단, 최소 횟수의 단위 동작을 사용해야 한다.  
  
### 입력  
첫째 줄에 주어진 평지의 한 변의 길이 N이 주어진다. (4<=N<=50) 주어진다. 이어서 그 지형의 정보가 0, 1, B, E로 이루어진 문자열로 주어진다. 한 줄에 입력되는 문자열의 길이는 N이며 입력 문자 사이에는 빈칸이 없다. 통나무와 최종 위치의 개수는 1개이다.  
  
### 출력  
첫째 줄에 최소 동작 횟수를 출력한다. 이동이 불가능하면 0만을 출력한다.  

```python
from queue import PriorityQueue

def check_possible(N, board, checked, y, x, is_vertical, command): # 명령 수행 가능한지 체크
    if command == 'T': # 90도 회전
        for add_y in range(-1, 2):
            for add_x in range(-1, 2):
                if not(0 <= add_y+y < N and 0 <= add_x+x < N ) or board[add_y+y][add_x+x] == '1':
                    return False
        # 통나무를 포함하는 주번 3X3칸이 비었는 지 확인
        return True
    elif command == 'U': # 위로 이동
        if is_vertical: # 세로
            if y-2 < 0 or board[y-2][x] == '1':
                return False
        else: # 가로
            if y-1 < 0:
                return False
            for add_x in range(-1, 2):
                if board[y-1][x+add_x] == '1':
                    return False
        # 이동할 칸이 비었는지 확인
        return True
    elif command == 'D': # 아래로 이동
        if is_vertical: # 세로
            if y+2 > N-1 or board[y+1][x] == '1':
                return False
        else: # 가로
            if y+1 > N-1:
                return False
            for add_x in range(-1, 2):
                if board[y+1][x+add_x] == '1':
                    return False
        # 이동할 칸이 비었는지 확인
        return True
    elif command == 'L': # 왼쪽으로 이동
        if is_vertical: # 세로
            if x-1 < 0:
                return False
            for add_y in range(-1, 2):
                if board[y+add_y][x-1] == '1':
                    return False
        else: # 가로
            if x-2 < 0 or board[y][x-2] == '1':
                return False
        # 이동할 칸이 비었는지 확인
        return True
    elif command == 'R': # 오른쪽으로 이동
        if is_vertical: # 세로
            if x+1 > N-1:
                return False
            for add_y in range(-1, 2):
                if board[y+add_y][x+1] == '1':
                    return False
        else: # 가로
            if x+2 > N-1 or board[y][x+2] == '1':
                return False
        # 이동할 칸이 비었는지 확인
        return True
        
def simulation(N, board, checked, log_y, log_x, log_vertical, arrival_y, arrival_x, arrival_vertical):
    commands = [(-1, 0, 'U'), (1, 0, 'D'), (0, -1,'L'), (0, 1, 'R'), (0, 0, 'T')]
    # 명령들 U. D. L. R. T
    q = PriorityQueue() # 우선순위 큐
    q.put((0, log_y, log_x, log_vertical)) # count, 중심 y좌표, 중심 x좌표, 통나무가 세로로 서있는지 여부
    while not q.empty():
        count, y, x, is_vertical = q.get() # count, 중심 y좌표, 중심 x좌표, 통나무가 세로로 서있는지 여부
        d_idx = 0
        if is_vertical:
            d_idx = 1
        # 통나무가 서있는 방향 인덱스
        if y==arrival_y and x==arrival_x and is_vertical==arrival_vertical: # 통나무가 원하는 상태에 도달했을 때
            return count # 명령 횟수 반환
        for add_y, add_x, command in commands: # 모든 명령에 대해서
            if check_possible(N, board, checked, y, x, is_vertical, command): # 명령 수행이 가능하다면
                if command == 'T': # 90도 회전
                    if is_vertical and checked[y][x][0] > count+1:
                        q.put((count+1, y, x, False))
                    elif not is_vertical and checked[y][x][1] > count+1:
                        q.put((count+1, y, x, True))
                elif command == 'U' and checked[y-1][x][d_idx] > count+1: # 위로 이동
                    q.put((count+1, y-1, x, is_vertical))
                elif command == 'D' and checked[y+1][x][d_idx] > count+1: # 아래로 이동
                    q.put((count+1, y+1, x, is_vertical))
                elif command == 'L' and checked[y][x-1][d_idx] > count+1: # 왼쪽으로 이동
                    q.put((count+1, y, x-1, is_vertical))
                elif command == 'R' and checked[y][x+1][d_idx] > count+1: # 오른쪽으로 이동
                    q.put((count+1, y, x+1, is_vertical))
    return 0

N = int(input()) # 범위 크기
board = [] # 보드
for _ in range(N):
    board.append(list(input()))
log_y, log_x = 0, 0 # 통나무의 중심 좌표 y,x
log_vertical = None # 통나무가 서있는 상태(세로, 가로)
log_num = 0
arrival_y, arrival_x = 0, 0 # 도착지의 중심 좌표 y,x
arrival_vertical = None # 도착지가 서있는 상태
arrival_num = 0
checked = [[[pow(N-1,2) for _ in range(2)] for _ in range(N)] for _ in range(N)]
# 상태에 따른 명령 이행 최소 개수

for y in range(N):
    for x in range(N):
        if board[y][x] == 'B': # 통나무에 해당하는 좌표
            board[y][x] = '0'
            if log_num == 1:
                if log_y == y:
                    log_vertical = False
                else:
                    log_vertical = True
            log_y += y
            log_x += x
            log_num += 1
        if board[y][x] == 'E': # 도착지에 해당하는 좌표
            board[y][x] = '0'
            if arrival_num == 1:
                if arrival_y == y:
                    arrival_vertical = False
                else:
                    arrival_vertical = True
            arrival_y += y
            arrival_x += x
            arrival_num += 1
log_y //= 3
log_x //= 3
arrival_y //= 3
arrival_x //= 3
if log_vertical:
    checked[log_y][log_x][1] = 0
else:
    checked[log_y][log_x][0] = 0
# 처음 통나무 상태에 대한 명령개수 갱신
print(simulation(N, board, checked, log_y, log_x, log_vertical, arrival_y, arrival_x, arrival_vertical))
# 답 출력
```

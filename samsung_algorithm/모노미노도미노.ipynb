<그림 14>
  
행이나 열이 타일로 가득찬 경우와 연한 칸에 블록이 있는 경우가 동시에 발생할 수 있다. 이 경우에는 행이나 열이 타일로 가득 찬 경우가 없을 때까지 점수를 획득하는 과정이 모두 진행된 후, 연한 칸에 블록이 있는 경우를 처리해야 한다.  
  
블록은 보드에 놓인 이후에 다른 블록과 합쳐지지 않는다. 나누어지는 경우는 행이나 열이 타일로 가득찬 경우에 발생할 수 있다. 초록색 보드에서는 2×1 블록에서 일부가 사라져서 1×1 블록이 될 수 있고, 파란색 보드에서는 1×2 블록에서 일부가 사라져서 1×1 블록이 될 수 있다.  
  
블록을 놓은 위치가 순서대로 주어졌을 때, 얻은 점수와 초록색 보드와 파란색 보드에 타일이 있는 칸의 개수를 모두 구해보자.  
  
### 입력  
첫째 줄에 블록을 놓은 횟수 N(1 ≤ N ≤ 10,000)이 주어진다.  
  
둘째 줄부터 N개의 줄에 블록을 놓은 정보가 한 줄에 하나씩 순서대로 주어지며, t x y와 같은 형태이다.  
  
- t = 1: 크기가 1×1인 블록을 (x, y)에 놓은 경우  
- t = 2: 크기가 1×2인 블록을 (x, y), (x, y+1)에 놓은 경우  
- t = 3: 크기가 2×1인 블록을 (x, y), (x+1, y)에 놓은 경우  
  
블록이 차지하는 칸이 빨간색 칸의 경계를 넘어가는 경우는 입력으로 주어지지 않는다.  
  
### 출력  
첫째 줄에 블록을 모두 놓았을 때 얻은 점수를 출력한다.  
둘째 줄에는 파란색 보드와 초록색 보드에서 타일이 들어있는 칸의 개수를 출력한다.  
   

```python
from pprint import pprint

def collocate_block(t, y, x): # 블록을 배치
    global blue_board
    global green_board
    global blue_num
    global green_num
        
    if t == 1: # 1x1 블록 배치
        green_check = False
        blue_check = False
        loc_green = 5
        loc_blue = 5
        for i in range(1, 6):
            if not green_check and green_board[i][x]:
                loc_green = i-1
            if not blue_check and blue_board[i][y]:
                loc_blue = i-1
            if green_check and blue_check:
                break
            # 해당 위치에서 블록을 배치할 때 배치되어야할 위치 확인
        green_board[loc_green][x] = True
        green_num[x] += 1
        blue_board[loc_blue][y] = True
        blue_num[y] += 1
        # 배치된 부분 체크 및 파란색 초록색 부분의 배치된 수 증가
        
    elif t == 2: # 1x2 블록 배치
        green_check = False
        green2_check = False
        blue_check = False
        loc_green = 5
        loc_blue = 5
        for i in range(1, 6):
            if not green_check and green_board[i][x]:
                loc_green = min(loc_green, i-1)
                green_check = True
            if not green2_check and green_board[i][x+1]:
                loc_green = min(loc_green, i-1)
                green2_check = True
            if not blue_check and blue_board[i][y]:
                loc_blue = i-1
                blue_check = True
            if green_check and green2_check and blue_check:
                break
            # 해당 위치에서 블록을 배치할 때 배치되어야할 위치 확인
        green_board[loc_green][x] = True
        green_num[x] += 1
        green_board[loc_green][x+1] = True
        green_num[x+1] += 1
        blue_board[loc_blue][y] = True
        blue_board[loc_blue-1][y] = True
        blue_num[y] += 2
        # 배치된 부분 체크 및 파란색 초록색 부분의 배치된 수 증가
        
    elif t == 3: # 2x1 블록 배치
        green_check = False
        blue_check = False
        blue2_check = False
        loc_green = 5
        loc_blue = 5
        for i in range(1, 6):
            if not green_check and green_board[i][x]:
                loc_green = i-1
                green_check = True
            if not blue_check and blue_board[i][y]:
                loc_blue = min(loc_blue, i-1)
                blue_check = True
            if not blue2_check and blue_board[i][y+1]:
                loc_blue = min(loc_blue, i-1)
                blue2_check = True
            if green_check and blue_check and blue2_check:
                break
            # 해당 위치에서 블록을 배치할 때 배치되어야할 위치 확인
        green_board[loc_green-1][x] = True
        green_board[loc_green][x] = True
        green_num[x] += 2
        blue_board[loc_blue][y] = True
        blue_num[y] += 1
        blue_board[loc_blue][y+1] = True
        blue_num[y+1] += 1
        # 배치된 부분 체크 및 파란색 초록색 부분의 배치된 수 증가
    sorting_block() # 배치 한 후의 상황에서 다시 블록을 정렬
    return

def sorting_block(): # 블록을 정렬하는 함수
    global blue_board
    global green_board
    global blue_num
    global green_num
    global point
    
    changed = False
    
    blue_moved = False
    green_moved = False
    blue_matched = False
    green_matched = False
    for i in range(0, 2): # 옅을 색의 블록 구역에서
        if not blue_moved and True in blue_board[i]: # 파란색 블록쪽에 차지하는 블록이 있다면
            blue_moved = True
            changed = True
            for j in range(6-(2-i), 6):
                for k in range(4):
                    if blue_board[j][k]:
                        blue_num[k] -= 1 # 없어지는 블록들만큼 값들을 재설정
            for j in range(5, 1, -1):
                blue_board[j] = blue_board[j-(2-i)]
                # 그리고 파란색 블록의 열을 오른쪽으로 옮긴다.
            for j in range(i, 2):
                blue_board[j] = [False for _ in range(4)]
                # 옮긴다음 빈 부분은 빈칸으로 다시 설정한다.
        if not green_moved and True in green_board[i]: # 초록색 블록쪽에 차지하는 블록이 있다면
            green_moved = True
            changed = True
            for j in range(6-(2-i), 6):
                for k in range(4):
                    if green_board[j][k]:
                        green_num[k] -= 1 # 없어지는 블록들만큼 값들을 재설정
            for j in range(5, 1, -1):
                green_board[j] = green_board[j-(2-i)]
                # 그리고 초록색 블록의 행을 아래쪽으로 옮긴다.
            for j in range(i, 2):
                green_board[j] = [False for _ in range(4)]
                # 옮긴다음 빈 부분은 빈칸으로 다시 설정한다.
                
    for i in range(2, 6): # 만약 한 줄(행 및 열)이 완성되었다면
        if False in blue_board[i] or blue_matched:
            pass
        else: # 파란색 블록에서의 한 열이 완성되었을 때
            changed = True
            blue_matched = True
            point += 1
            blue_board = [[False for _ in range(4)] for _ in range(6)]
            for j in range(4):
                blue_num[j] -= 1
                for k in range(5, 5-blue_num[j], -1):
                    blue_board[k][j] = True
                # 파란색 블록에서의 모든 블록들을 오른쪽으로 땡긴다.
        if False in green_board[i] or green_matched:
            pass
        else: # 초록색 블록에서의 한 행이 완성되었을 때
            changed = True
            green_matched = True
            point += 1
            green_board = [[False for _ in range(4)] for _ in range(6)]
            for j in range(4):
                green_num[j] -= 1
                for k in range(5, 5-green_num[j], -1):
                    green_board[k][j] = True
                # 초록색 블록에서의 모든 블록들을 아래쪽으로 땡긴다.
    if changed: # 만약 변화가 있었다면 다시 한번 재귀적으로 재정렬해준다.
        sorting_block()
    return

while True:
    print("Input")
    N = int(input("N: ")) # 블록 배치 개수
    if N == 0:
        print("Exit the program")
        break
    point = 0 # 점수
    blue_num = [0 for _ in range(4)] # 파란색 블록에서 각 행당 블록 개수
    green_num = [0 for _ in range(4)] # 초록색 블록에서 각 열당 블록 개수
    blue_board = [[False for _ in range(4)] for _ in range(6)] # 파란색 블록
    green_board = [[False for _ in range(4)] for _ in range(6)] # 초록색 블록
    for _ in range(N):
        t, x, y = tuple(int(x) for x in input("t, x, y: ").split())
        # 블록 모양, x좌표, y좌표
        collocate_block(t, x, y) # 블록 배치
    #     print("===========================")
    #     pprint(green_board)
    #     pprint(blue_board)
    #     print("===========================")

    print("\nOutput")
    print(point) # 총 점수
    print(sum(blue_num) + sum(green_num)) # 파란색 블록과 초록색 블록에 배치된 블록의 총 개수
    print()
```

### 문제  
백준시의 시장 최백준은 지난 몇 년간 게리맨더링을 통해서 자신의 당에게 유리하게 선거구를 획정했다. 견제할 권력이 없어진 최백준은 권력을 매우 부당하게 행사했고, 심지어는 시의 이름도 백준시로 변경했다. 이번 선거에서는 최대한 공평하게 선거구를 획정하려고 한다.  
  
백준시는 N개의 구역으로 나누어져 있고, 구역은 1번부터 N번까지 번호가 매겨져 있다. 구역을 두 개의 선거구로 나눠야 하고, 각 구역은 두 선거구 중 하나에 포함되어야 한다. 선거구는 구역을 적어도 하나 포함해야 하고, 한 선거구에 포함되어 있는 구역은 모두 연결되어 있어야 한다. 구역 A에서 인접한 구역을 통해서 구역 B로 갈 수 있을 때, 두 구역은 연결되어 있다고 한다. 중간에 통하는 인접한 구역은 0개 이상이어야 하고, 모두 같은 선거구에 포함된 구역이어야 한다.  
  
아래 그림은 6개의 구역이 있는 것이고, 인접한 구역은 선으로 연결되어 있다.

공평하게 선거구를 나누기 위해 두 선거구에 포함된 인구의 차이를 최소로 하려고 한다. 백준시의 정보가 주어졌을 때, 인구 차이의 최솟값을 구해보자.  
  
### 입력
첫째 줄에 구역의 개수 N이 주어진다. 둘째 줄에 구역의 인구가 1번 구역부터 N번 구역까지 순서대로 주어진다. 인구는 공백으로 구분되어져 있다.  
  
셋째 줄부터 N개의 줄에 각 구역과 인접한 구역의 정보가 주어진다. 각 정보의 첫 번째 정수는 그 구역과 인접한 구역의 수이고, 이후 인접한 구역의 번호가 주어진다. 모든 값은 정수로 구분되어져 있다.  
  
구역 A가 구역 B와 인접하면 구역 B도 구역 A와 인접하다. 인접한 구역이 없을 수도 있다.  
  
### 출력  
첫째 줄에 백준시를 두 선거구로 나누었을 때, 두 선거구의 인구 차이의 최솟값을 출력한다. 두 선거구로 나눌 수 없는 경우에는 -1을 출력한다.  
  
### 제한  
- 2 ≤ N ≤ 10
- 1 ≤ 구역의 인구 수 ≤ 100  
  

```python
from queue import Queue
from copy import deepcopy

def search_district(N, population, adjacent, po_diff, total): 
    # 각 지역별로 시작에서 DFS를 통해 선거구 지역을 늘려가면서 두 선거구로 나눠졌을 때 최소 인구차이를 구하고 반환 
    answer = total # 답 (모든 지역의 총 인구수로 초기화)
    all_included = pow(2, N) - 1 # (비트마스크) N개의 모든 지역구가 포함되었을 경우의 값
    pop = [-1 for _ in range(pow(2, N))] # 지역구의 상황에 따른 인구수
    for i in range(N): # 모든 지역에 대해서
        q = Queue() # 큐
        pop[pow(2,i)] = population[i] # i번 지역만을 포함하는 선거구의 인구는 i번 지역의 인구이다
        candidate = set() # 선거구 포함할 후보 지역
        for j in range(N):
            if adjacent[i][j] == 1: # i번 지역과 인접한 지역들에 대해서
                candidate.add(j) # 선거구 확장 후보지에 추가
        q.put([pow(2,i), candidate]) # 큐에 [현재 선거구 상태, 선거구 확장 후보지] 정보를 push 
        while not q.empty(): # 큐가 빌때까지
            status, candidate = q.get() # 큐에서 [현재 선거구 상태, 선거구 확장 후보지] 정보를 pop
            if pop[all_included - status] != -1: # 만약 현재 탐색중인 선거구를 제외한 지역 또한 하나의 선거구로 묶일 수 있을 때
                answer = min(answer, abs(pop[status] - pop[all_included - status]))
                # 이는 전체 선거구가 2개의 선거구로 나눠질 수 있는 경우이므로 인구 수 차이의 최솟값을 저장
            for c in candidate: # 모든 선거구 확장 후보지에 대해서
                if pop[status + pow(2, c)] != -1: # 이미 탐색한 선거구 상태라면 넘어간다
                    continue
                cp_candidate = deepcopy(candidate)
                cp_candidate.remove(c) # 확장할 c번 지역은 후보지에서 제거하고,
                for k in range(N): # c번 지역과 인접하고, 이미 포함된 지역을 제외한 지역에 대해서
                    if adjacent[c][k] == 1 and status & pow(2, k) == 0:
                        cp_candidate.add(k) # 확장 후보지에 추가해준다.
                pop[status + pow(2, c)] = pop[status] + population[c] # c번 지역을 포함한 선거구에 대해서 인구값을 갱신
                q.put([status+pow(2, c), cp_candidate]) # [c번 지역을 포함한 선거구 상태, 추가된 확장 후보지] 정보를 다시 큐에 push 
    return answer # 답 반환



T = int(input("Testcase num: ")) # 테스트케이스 개수
for _ in range(T):
    print("\n\nInput")
    N = int(input()) # 각 지역의 개수
    population = list(map(int, input().split())) # 각 지역별 인구
    total = sum(population) # 인구 총 합
    adjacent = [[0 for _ in range(N)] for _ in range(N)] # 인접행렬
    po_diff = [[-1 for _ in range(pow(2, N))] for _ in range(pow(2, N))]
    # 각 선거구 설정에 따른 인구 차
    for i in range(N): # 모든 지역에 대해서
        inp = list(map(int, input().split()))[1:]
        for j in inp:
            adjacent[i][j-1] = 1 # 인접한 지역끼리는 1의 값을 저장
    answer = search_district(N, population, adjacent, po_diff, total) # 시뮬레이션을 통해 최소 인구차를 보일 수 있는 선거구 획정
    if answer == total: # 만약 어떠한 선거구도 둘로 나눠지지 않았을 경우
        answer = -1 # -1이 답이 된다.
    print("\nOutput")
    print(answer) # 최소 인구차 출력
    
```

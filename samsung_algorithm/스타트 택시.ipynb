# 스타트 택시

[baekjoon의 '스타트 택시' 링크](https://www.acmicpc.net/problem/19238)

### 문제  
스타트링크가 "스타트 택시"라는 이름의 택시 사업을 시작했다. 스타트 택시는 특이하게도 손님을 도착지로 데려다줄 때마다 연료가 충전되고, 연료가 바닥나면 그 날의 업무가 끝난다.  
  
택시 기사 최백준은 오늘 M명의 승객을 태우는 것이 목표이다. 백준이 활동할 영역은 N×N 크기의 격자로 나타낼 수 있고, 각 칸은 비어 있거나 벽이 놓여 있다. 택시가 빈칸에 있을 때, 상하좌우로 인접한 빈칸 중 하나로 이동할 수 있다. 알고리즘 경력이 많은 백준은 특정 위치로 이동할 때 항상 최단경로로만 이동한다.  
  
M명의 승객은 빈칸 중 하나에 서 있으며, 다른 빈칸 중 하나로 이동하려고 한다. 여러 승객이 같이 탑승하는 경우는 없다. 따라서 백준은 한 승객을 태워 목적지로 이동시키는 일을 M번 반복해야 한다. 각 승객은 스스로 움직이지 않으며, 출발지에서만 택시에 탈 수 있고, 목적지에서만 택시에서 내릴 수 있다.  
  
백준이 태울 승객을 고를 때는 현재 위치에서 최단거리가 가장 짧은 승객을 고른다. 그런 승객이 여러 명이면 그중 행 번호가 가장 작은 승객을, 그런 승객도 여러 명이면 그중 열 번호가 가장 작은 승객을 고른다. 택시와 승객이 같은 위치에 서 있으면 그 승객까지의 최단거리는 0이다. 연료는 한 칸 이동할 때마다 1만큼 소모된다. 한 승객을 목적지로 성공적으로 이동시키면, 그 승객을 태워 이동하면서 소모한 연료 양의 두 배가 충전된다. 이동하는 도중에 연료가 바닥나면 이동에 실패하고, 그 날의 업무가 끝난다. 승객을 목적지로 이동시킨 동시에 연료가 바닥나는 경우는 실패한 것으로 간주하지 않는다.

모든 승객을 성공적으로 데려다줄 수 있는지 알아내고, 데려다줄 수 있을 경우 최종적으로 남는 연료의 양을 출력하는 프로그램을 작성하시오.  
  
### 입력  
첫 줄에 N, M, 그리고 초기 연료의 양이 주어진다. (2 ≤ N ≤ 20, 1 ≤ M ≤ N2, 1 ≤ 초기 연료 ≤ 500,000) 연료는 무한히 많이 담을 수 있기 때문에, 초기 연료의 양을 넘어서 충전될 수도 있다.  
  
다음 줄부터 N개의 줄에 걸쳐 백준이 활동할 영역의 지도가 주어진다. 0은 빈칸, 1은 벽을 나타낸다.  
  
다음 줄에는 백준이 운전을 시작하는 칸의 행 번호와 열 번호가 주어진다. 행과 열 번호는 1 이상 N 이하의 자연수이고, 운전을 시작하는 칸은 빈칸이다.  
  
그다음 줄부터 M개의 줄에 걸쳐 각 승객의 출발지의 행과 열 번호, 그리고 목적지의 행과 열 번호가 주어진다. 모든 출발지와 목적지는 빈칸이고, 모든 출발지는 서로 다르며, 각 손님의 출발지와 목적지는 다르다.  
  
### 출력  
모든 손님을 이동시키고 연료를 충전했을 때 남은 연료의 양을 출력한다. 단, 이동 도중에 연료가 바닥나서 다음 출발지나 목적지로 이동할 수 없으면 -1을 출력한다. 모든 손님을 이동시킬 수 없는 경우에도 -1을 출력한다.  

```python
from queue import PriorityQueue

def calculate_min(N, M, board, depart, dest): # depart부터 dest까지의 최소 거리 반환
    visited = [[N*M for _ in range(N)] for _ in range(N)] # 해당 위치로 가기 위한 최소 이동 횟수
    s_y, s_x = depart # 출발 위치
    e_y, e_x = dest # 도착 위치
    q = PriorityQueue() # 우선순위 큐
    q.put((0, s_y, s_x)) # (이동거리, 현재 y, 현재 x) 정보를 큐에 push
    direction = [(0,1), (1,0), (0,-1), (-1,0)] # 인접한 4방향
    while not q.empty():
        count, y, x = q.get() # (이동거리, 현재 y, 현재 x) 정보를 큐에서 pop
        if y == e_y and x == e_x: # 만약 목적지에 도달했다면 이동거리 반환
            return count
        for add_y, add_x in direction: # 인접한 네 방향에 대해서
            if 0 <= add_y+y < N and 0 <= add_x+x < N and board[add_y+y][add_x+x] == 0 and visited[add_y+y][add_x+x] > count+1:
            # 보드 상의 빈칸이라면
                visited[add_y+y][add_x+x] = count+1
                q.put((count+1, add_y+y, add_x+x)) # (이동거리+1, 이동 후 y, 이동 후 x) 정보를 큐에 push
    return -1

def simulation(N, M, fuel, board, s_y, s_x, guest_info):
    y, x = s_y, s_x # 현재 위치
    visited = [False for _ in range(M)] # 손님 방문 여부
    while True:
        distance = N*N # 거리
        idx = 0 # 인덱스
        visited_num = 0 # 이미 태운 손님수
        for i in range(M): # 모든 손님에 대해서
            
            if visited[i]: # 만약 이미 태운 손님일 때
                visited_num += 1 # 이미 태운 손님 수 1 증가
                continue # 다음 탐색으로 넘어감  
            curr_dist = calculate_min(N, M, board, (y, x), guest_info[i][:2]) # 현재 위치에서 손님까지의 최소 거리
            if curr_dist == -1: # 도달할 수 없는 손님이 있다면 -1 반환
                return -1
            if distance > curr_dist: # 만약 최소 거리라면
                distance = curr_dist # 거리 갱신
                idx = i # 현재 태우러갈 손님 인덱스 갱신
                
        if visited_num == M: # 만약 모든 손님을 다 태웠다면
            return fuel # 현재 연료 상태 반환
        
        if fuel - distance < 0: # 손님을 태우러 가는 도중에 연료가 떨어졌다면 -1 반환
            return -1
        
        visited[idx] = True # idx번째 손님을 태웠음을 체크
        
        fuel -= distance # 손님을 태우러 가기 위해 사용한 연료만큼 감소
        y, x = guest_info[idx][:2] # 손님의 위치가 현재 위치가 된다.
        distance = calculate_min(N, M, board, guest_info[idx][:2], guest_info[idx][2:]) # 손님을 태워서 목적지까지 가기 위한 최소 거리
        if distance == -1: # 도달할 수 없는 곳에 목적지가 있다면 -1 반환
            return -1 
        if fuel - distance < 0: # 만약 목적지까지 가기전에 연료가 떨어진다면 -1 반환
            return -1
        fuel += distance # 목적지 도달 후 태운 거리 * 2만큼 충전되므로 더해준다.
        y, x = guest_info[idx][2:] # 현재 위치를 목적지 위치로 갱신
    return -1
    

T = int(input("Testcase: ")) # 테스트 케이스 수
for _ in range(T):
    print("\nInput")
    N, M, fuel = tuple(int(x) for x in input().split()) # 보드 크기, 손님 수, 초기 연료
    board = [] # 보드
    for _ in range(N):
        board.append(list(int(x) for x in input().split()))
    s_y, s_x = tuple(int(x)-1 for x in input().split()) # 현재 위치
    guest_info = [] # 손님 정보
    for _ in range(M):
        guest_info.append(tuple(int(x)-1 for x in input().split()))
    guest_info.sort() # 손님을 행,열 오름차순으로 정렬
    print("\nOutput")
    print(simulation(N, M, fuel, board, s_y, s_x, guest_info)) # 시뮬레이션 진행 후 답 출력
```

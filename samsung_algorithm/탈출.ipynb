# 탈출

[baekjoon의 '탈출' 링크](https://www.acmicpc.net/problem/3055)

### 문제   
사악한 암흑의 군주 이민혁은 드디어 마법 구슬을 손에 넣었고, 그 능력을 실험해보기 위해 근처의 티떱숲에 홍수를 일으키려고 한다. 이 숲에는 고슴도치가 한 마리 살고 있다. 고슴도치는 제일 친한 친구인 비버의 굴로 가능한 빨리 도망가 홍수를 피하려고 한다.  
  
티떱숲의 지도는 R행 C열로 이루어져 있다. 비어있는 곳은 '.'로 표시되어 있고, 물이 차있는 지역은 '\*', 돌은 'X'로 표시되어 있다. 비버의 굴은 'D'로, 고슴도치의 위치는 'S'로 나타내어져 있다.  
  
매 분마다 고슴도치는 현재 있는 칸과 인접한 네 칸 중 하나로 이동할 수 있다. (위, 아래, 오른쪽, 왼쪽) 물도 매 분마다 비어있는 칸으로 확장한다. 물이 있는 칸과 인접해있는 비어있는 칸(적어도 한 변을 공유)은 물이 차게 된다. 물과 고슴도치는 돌을 통과할 수 없다. 또, 고슴도치는 물로 차있는 구역으로 이동할 수 없고, 물도 비버의 소굴로 이동할 수 없다.  
  
티떱숲의 지도가 주어졌을 때, 고슴도치가 안전하게 비버의 굴로 이동하기 위해 필요한 최소 시간을 구하는 프로그램을 작성하시오.  
  
고슴도치는 물이 찰 예정인 칸으로 이동할 수 없다. 즉, 다음 시간에 물이 찰 예정인 칸으로 고슴도치는 이동할 수 없다. 이동할 수 있으면 고슴도치가 물에 빠지기 때문이다.  
  
### 입력  
첫째 줄에 50보다 작거나 같은 자연수 R과 C가 주어진다.  
  
다음 R개 줄에는 티떱숲의 지도가 주어지며, 문제에서 설명한 문자만 주어진다. 'D'와 'S'는 하나씩만 주어진다.  
  
### 출력  
첫째 줄에 고슴도치가 비버의 굴로 이동할 수 있는 가장 빠른 시간을 출력한다. 만약, 안전하게 비버의 굴로 이동할 수 없다면, "KAKTUS"를 출력한다.  

```python
from queue import PriorityQueue

def water_move(R, C, board, water, e_y, e_x, count): # 이동시간에 따른 물이 고인 위치
    if count < len(water): # 만약 이미 계산이 끝난 경우 위치 탐색 종료
        return
    new_area = set() # 새로 물이 고일 위치
    new_area.union(water[count-1]) # 현재 물이 고인 위치들도 포함한다.
    direction = [(0,1), (1,0), (0,-1), (-1,0)] # 4방향
    for y, x in water[count-1]: # 현재 물이 고인 위치들에 대해서
        for add_y, add_x in direction: # 4방향으로 탐색
            if 0 <= add_y+y < R and 0 <= add_x+x < C and board[add_y+y][add_x+x] != 'X' and not(add_y+y == e_y and add_x+x == e_x):
            # 보드 상의 위치이고, 벽이나 비버굴의 위치를 제외한 위치일 경우
                new_area.add((y+add_y, x+add_x))
                # 새로 물이 찰 지역으로 추가한다.
    water.append(new_area) # 그리고 이동시간에 따른 물이 고인 위치를 추가
    return
    

def simulation(R, C, board, water, checked, s_y, s_x, e_y, e_x): # 고슴도치가 비버의 굴로 이동하는 최소 시간을 구하는 시뮬레이션
    q = PriorityQueue() # 우선순위 큐(최소시간상황부터 탐색)
    direction = [(0,1), (1,0), (0,-1), (-1,0)] # 4방향
    q.put((0, s_y, s_x)) # 현재흐른시간 0초, 시작 좌표 y, 시작 좌표 x 
    while not q.empty():
        count, y, x = q.get() # 현재 흐른 시간, 현재 좌표 y, 현재 좌표 x를 큐에서 pop
        if y == e_y and x == e_x: # 만약 비버 굴에 도달 했다면
            return count # 현재 흐른 시간 반환
        water_move(R, C, board, water, e_y, e_x, count+1) # 1초 후의 물이 고인 위치 탐색
        water_area = water[count+1] # 1초 후의 물이 고인 위치
        for add_y, add_x in direction: # 4방향에 대해서
            if 0 <= add_y+y < R and 0 <= add_x+x < C and board[add_y+y][add_x+x] != 'X' and checked[add_y+y][add_x+x] > count+1:
            # 보드 상의 벽의 위치를 제외한 위치에 대해서
                if (add_y+y, add_x+x) in water_area: # 물이 고인 위치일 경우 제외
                    continue
                checked[add_y+y][add_x+x] = count+1
                q.put((count+1, add_y+y, add_x+x)) # 이동하고 1초를 더해서 다시 큐에 push
    return -1 # 만약 도달하지 못한다면 -1 반환

T = int(input("TestCase: ")) # 테스트케이스 개수
while T > 0:
    print()
    print("Input")
    R, C = tuple(int(x) for x in input().split()) # 행, 열
    board = [] # 보드
    water = [] # 흐른 시간별 물이 채워진 위치
    checked = [[R*C for _ in range(C)] for _ in range(R)]
    water.append(set())
    s_y, s_x = 0, 0 # 처음 고슴도치의 위치 좌표
    e_y, e_x = 0, 0 # 비버굴의 위치 좌표
    for y in range(R):
        row = list(input())
        for x in range(C):
            if row[x] == '*': # 물
                water[0].add((y, x)) # 0초가 지났을 때의 물의 위치에 포함
                row[x] = '.'
            elif row[x] == 'D': # 비버굴
                e_y, e_x = y, x # 비버굴의 위치 갱신
                row[x] = '.'
            elif row[x] == 'S': # 고슴도치
                checked[y][x] = 0 # 고슴도치가 0초에 (y, x)에 도달해있음을 체크
                s_y, s_x = y, x # 고슴도치의 위치 갱신
                row[x] = '.'
        board.append(row)
    print("\nOutput")
    answer = simulation(R, C, board, water, checked, s_y, s_x, e_y, e_x)
    # 고슴도치가 비버굴까지 도달할 수 있는 최소 시간
    if answer == -1: # 만약 도달 불가할 경우
        print("KAKTUS") # "KAKTUS" 출력
    else: # 도달 가능할 경우
        print(answer) # 최소 시간 출력
    T -= 1
```

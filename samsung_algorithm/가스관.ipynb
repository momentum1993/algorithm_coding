# 가스관

[baekjoon의 '가스관' 링크](https://www.acmicpc.net/problem/2931)

### 문제   
러시아 가스를 크로아티아로 운반하기 위해 자그레브와 모스코바는 파이프라인을 디자인하고 있다. 두 사람은 실제 디자인을 하기 전에 파이프 매니아 게임을 이용해서 설계를 해보려고 한다.  
  
이 게임에서 유럽은 R행 C열로 나누어져 있다. 각 칸은 비어있거나, 아래 그림과 같은 일곱가지 기본 블록으로 이루어져 있다.  
  
![](https://www.acmicpc.net/upload/images/pipe1.png)  
  
가스는 모스크바에서 자그레브로 흐른다. 가스는 블록을 통해 양방향으로 흐를 수 있다. '+'는 특별한 블록으로, 아래 예시처럼 두 방향 (수직, 수평)으로 흘러야 한다.  
  
![](https://www.acmicpc.net/upload/images/pipe2(1).png)  
  
파이프 라인의 설계를 마친 후 두 사람은 잠시 저녁을 먹으러 갔다. 그 사이 해커가 침임해 블록 하나를 지웠다. 지운 블록은 빈 칸이 되어있다.  
  
해커가 어떤 칸을 지웠고, 그 칸에는 원래 어떤 블록이 있었는지 구하는 프로그램을 작성하시오.  
  
### 입력  
첫째 줄에 유럽의 크기 R과 C가 주어진다. (1 ≤ R, C ≤ 25)  
  
다음 R개 줄에는 C개 글자가 주어지며, 다음과 같은 글자로 이루어져 있다.  
  
- 빈칸을 나타내는 '.'  
- 블록을 나타내는 '|'(아스키 124), '-','+','1','2','3','4'  
- 모스크바의 위치를 나타내는 'M'과 자그레브를 나타내는 'Z'. 두 글자는 한 번만 주어진다.  
  
항상 답이 존재하고, 가스의 흐름이 유일한 경우만 입력으로 주어진다, 또, 모스크바와 자그레브가 하나의 블록과 인접해 있는 입력만 주어진다. 또, 불필요한 블록이 존재하지 않는다. 즉, 없어진 블록을 추가하면, 모든 블록에 가스가 흐르게 된다.  
  
### 출력  
지워진 블록의 행과 열 위치를 출력하고, 어떤 블록이었는지를 출력한다.  

```python
def get_block(R, C, board, y, x):
    count = 0
    if 0 <= x+1 < C and board[y][x+1] in ['-','+','3','4']: # 동
        count += 1 << 3
    if 0 <= y+1 < R and board[y+1][x] in ['|','+','2','3']: # 남
        count += 1 << 2
    if 0 <= x-1 < C and board[y][x-1] in ['-','+','1','2']: # 서
        count += 1 << 1
    if 0 <= y-1 < R and board[y-1][x] in ['|','+','1','4']: # 북
        count += 1
        
    if count == 5: # 남, 북이 연결된 구조
        return '|'
    elif count == 10: # 동, 서가 연결된 구조
        return '-'
    elif count == 12: # 동, 남이 연결된 구조
        return '1'
    elif count == 9: # 동, 북이 연결된 구조
        return '2'
    elif count == 3: # 서, 북이 연결된 구조
        return '3'
    elif count == 6: # 서, 남이 연결된 구조
        return '4'
    elif count == 15: # 동, 서, 남, 북이 연결된 구조
        return '+'
    
    
    
def gas_flow(R, C, board, y, x, before=-1):
    if not(0 <= y < R and 0 <= x < C): # 범위를 벗어나면 탐색 중지
        return None
    answer = None # 답
    tmp_answer = None
    
    direction = [(0,1), (1,0), (0,-1), (-1,0)] # 동, 남, 서, 북
    if board[y][x] == '.': # 현재 빈칸을 탐색한다면(=해커가 지운 칸)
        block = get_block(R, C, board, y, x) # 지운 블록 탐색하고
        return y+1, x+1, block # y, x, 지운블록 반환
    elif board[y][x] == '|' or board[y][x] == '-' or board[y][x] == '+': # 블록 '|' or 블록 '-' or 블록 '+' 
        add_y, add_x = direction[before]
        tmp_answer = gas_flow(R, C, board, add_y+y, add_x+x, before)
        # 이동방향 그대로 탐색
    elif board[y][x] == '1': # 블록 1
        if before == 2: # 서쪽으로 이동 중이었을 때
            tmp_answer = gas_flow(R, C, board, y+1, x, 1) # 남쪽으로 이동해서 탐색
        elif before == 3: # 북쪽으로 이동 중이었을 때
            tmp_answer = gas_flow(R, C, board, y, x+1, 0) # 동쪽으로 이동해서 탐색
    elif board[y][x] == '2': # 블록 2
        if before == 1: # 남쪽으로 이동 중이었을 때
            tmp_answer = gas_flow(R, C, board, y, x+1, 0) # 동쪽으로 이동해서 탐색
        elif before == 2: # 서쪽으로 이동 중이었을 때
            tmp_answer = gas_flow(R, C, board, y-1, x, 3) # 북쪽으로 이동해서 탐색
    elif board[y][x] == '3': # 블록 3
        if before == 0: # 동쪽으로 이동 중이었을 때
            tmp_answer = gas_flow(R, C, board, y-1, x, 3) # 북쪽으로 이동해서 탐색
        elif before == 1: # 남쪽으로 이동 중이었을 때
            tmp_answer = gas_flow(R, C, board, y, x-1, 2) # 서쪽으로 이동해서 탐색
    elif board[y][x] == '4': # 블록 4
        if before == 0: # 동쪽으로 이동 중이었을 때
            tmp_answer = gas_flow(R, C, board, y+1, x, 1) # 남쪽으로 이동해서 탐색
        elif before == 3: # 북쪽으로 이동 중이었을 때
            tmp_answer = gas_flow(R, C, board, y, x-1, 2) # 서쪽으로 이동해서 탐색 
    elif board[y][x] == 'Z': # 자그레브
        return None # 자그레브에 도착했을 경우 잘못된 경로로 왔기 때문에 탐색 중지
    if tmp_answer:
        answer = tmp_answer
    return answer # 답 반환
    
T = int(input("Testcase num: ")) # 테스트 케이스 개수
while T > 0:
    print()
    print("Input\n")
    R, C = tuple(int(x) for x in input().split()) # 행, 열의 개수
    answer = None # 답
    board = [] # 보드
    m_y, m_x = 0, 0 # 모스크바의 위치 좌표
    for y in range(R):
        row = list(input())
        if 'M' in row:
            m_y = y
            m_x = row.index('M')
        board.append(row)
    direction = [(0,1), (1,0), (0,-1), (-1,0)] # 동, 남, 서, 북
    
    # 모스크바의 4방향에 대해서 길을 찾고 탐색
    for i in range(4):
        add_y, add_x = direction[i]
        if 0 <= m_y+add_y < R and 0 <= m_x+add_x < C and board[m_y+add_y][m_x+add_x] != '.':
            answer = gas_flow(R, C, board, m_y+add_y, m_x+add_x, i)
        if answer:
            break
    print("\nOutput\n")
    for a in answer:
        print(a, end=" ")
    # 답 출력 (y, x, 입력한 블럭)
    print("\n")
    T -= 1
```
